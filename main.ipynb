```python


df = pd.read_csv("atp_matches_2000_2019_merged.csv")  # or atp_matches_merged.csv
print(df.shape)
print(df.columns[:25])
```

```python
print(df.columns.tolist())

# quick search for rank-related columns
rank_cols = [c for c in df.columns if "rank" in c.lower()]
level_cols = [c for c in df.columns if "level" in c.lower()]
print("Rank columns:", rank_cols)
print("Level columns:", level_cols)
```

```python
print(df.columns[-10:])
```

```python
use_cols = [
    "surface", "round", "best_of",
    "winner_rank", "loser_rank",
    "winner_rank_points", "loser_rank_points"
]
df2 = df[use_cols].copy()
```

```python
df2 = df2.dropna(subset=["winner_rank", "loser_rank"])
```

```python
df2
```

```python

for c in ["winner_rank", "loser_rank", "winner_rank_points", "loser_rank_points"]:
    df2[c] = pd.to_numeric(df2[c], errors="coerce")

df2 = df2.dropna(subset=["winner_rank", "loser_rank"])

# Target: 1 if higher-ranked player won
df2["target"] = (df2["winner_rank"] < df2["loser_rank"]).astype(int)

# Helpful engineered features
df2["rank_gap"] = (df2["winner_rank"] - df2["loser_rank"]).abs()
df2["points_gap"] = (df2["winner_rank_points"] - df2["loser_rank_points"]).abs()

print(df2.shape)
print(df2["target"].value_counts(normalize=True))
df2.head()
```
